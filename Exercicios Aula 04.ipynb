```python
import pandas as pd
import geopy
import plotly.express as px
import ipywidgets as widgets
from matplotlib import pyplot as plt
from geopy.geocoders import Nominatim
from ipywidgets import fixed
from matplotlib import gridspec


data = pd.read_csv('datasets/kc_house_data.csv')
```

```python
# 1 Qual a média do preço de compra dos imóveis por "nível"?
# Nível 0 -> Preço entre R$ 0 e R$ 321.950
# Nível 1 -> Preço entre 321.950 e 450.000
# Nível 2 -> Preço entre 450.000 e 645.000
# Nível 3 -> Preço acima de 645.000
```

```python
for i in range ( len( data) ):
    
    if (data.loc[i, 'price'] > 0 ) & (data.loc[i, 'price'] < 321950 ):
        data.loc[i, 'nivel'] = 'nivel_0'
        
    elif (data.loc[i, 'price'] > 321950 ) & (data.loc[i, 'price'] < 450000 ):
        data.loc[i, 'nivel'] = 'nivel_1'

    elif (data.loc[i, 'price'] > 450000 ) & (data.loc[i, 'price'] < 645000 ):
        data.loc[i, 'nivel'] = 'nivel_2'

    elif (data.loc[i, 'price'] > 645000 ) :
        data.loc[i, 'nivel'] = 'nivel_3'
        
       
    
meanif0 = data.query('nivel == "nivel_0"')['price'].mean()
meanif1 = data.query('nivel == "nivel_1"')['price'].mean()
meanif2 = data.query('nivel == "nivel_2"')['price'].mean()
meanif3 = data.query('nivel == "nivel_3"')['price'].mean()


print(meanif0)
print(meanif1)
print(meanif2)
print(meanif3)
```

```python
# 2 Qual a média do tamanho da sala de estar dos imóveis por "size"?
# Size 0 -> Tamanho entre 0 e 1427 ft
# Size 1 -> Tamanho entre 1427 e 1910 ft
# Size 2 -> Tamanho entre 1910 e 2550 ft
# Size 3 -> Tamanho acima de 2550 ft
```

```python
for i in range ( len( data) ):
    
    if (data.loc[i, 'sqft_living'] > 0 ) & (data.loc[i, 'sqft_living'] < 1427 ):
        data.loc[i, 'size'] = 'size_0'
        
    elif (data.loc[i, 'sqft_living'] > 1427 ) & (data.loc[i, 'sqft_living'] < 1910 ):
        data.loc[i, 'size'] = 'size_1'

    elif (data.loc[i, 'sqft_living'] > 1910 ) & (data.loc[i, 'sqft_living'] < 2550 ):
        data.loc[i, 'size'] = 'size_2'

    elif (data.loc[i, 'sqft_living'] > 2550 ) :
        data.loc[i, 'size'] = 'size_3'
        
       
    
meanif0 = data.query('size == "size_0"')['sqft_living'].mean()
meanif1 = data.query('size == "size_1"')['sqft_living'].mean()
meanif2 = data.query('size == "size_2"')['sqft_living'].mean()
meanif3 = data.query('size == "size_3"')['sqft_living'].mean()


print(meanif0)
print(meanif1)
print(meanif2)
print(meanif3)
```

```python
# Create Empty Rows

data['place_id'] = 'NA'
data['osm_type'] = 'NA'
data['country'] = 'NA'
data['country_code'] = 'NA'

# Initialize Nominatim API

geolocator = Nominatim(user_agent='geoapiExercises')

for i in range( len (data ) ):
    print( 'Loop: {} / {}'.format(i, len( data ) ) )
    
    # Make Query
    
    query = str( data.loc[i, 'lat'] ) + ',' + str( data.loc[i, 'long'] )
    
    # API Request
    
    response = geolocator.reverse( query )
    
    # Populate Data
    
    
    data.loc[i, 'place_id' ] = response.raw['place_id']
        
    if 'osm_type' in response.raw['osm_type']:
        data.loc[i, 'osm_type' ] = response.raw['osm_type']
        
        data.loc[i, 'country' ] = response.raw['country']
    
        data.loc[i, 'country_code' ] = response.raw['country_code']
    
```

```python
# 4 Adicione os seguintes filtros no Mapa:

# Tamanho mínimo da área de sala de estar 'sqft_living'
# Número mínimo de banheiros  'bathrooms'
# Valor máximo dos preços     'price'
# Tamanho máximo da área do porão   'sqft_basement'
# Filtro das condições do imovel     'condition' criar coluna pra definir a condicao do imovel
# Filtro por ano de construção'      'yr_built'

# Biblioteca necessaria importada
```

```python
df = data

# Place Condition

df['condition_type'] = "regular"
df.loc[data['condition'] <= 2 , 'condition_type'] = 'bad'
df.loc[data['condition'] == 5 , 'condition_type'] = 'good'


# Iterative Buttons

sqft_min = widgets.IntSlider(
    value= 500,
    min = 290,
    max = 13540,
    step = 1,
    description='Minimum Size Living',
    disable=False,
    orientation='horizontal',
    readout=True ) 

bath_min = widgets.IntSlider(
    value= 1,
    min = 0,
    max = 30,
    step = 1,
    description='Minimum Bathroom Number',
    disable=False,
    orientation='horizontal',
    readout=True ) 

price_limit = widgets.IntSlider(
    value= 540000,
    min = 75000,
    max = 7700000,
    step = 1,
    description='Maximun Price',
    disable=False,
    orientation='horizontal',
    readout=True) 

basement_limit = widgets.IntSlider(
    value= 100,
    min = 0,
    max = 4820,
    step = 1,
    description='Maximun Size Basement',
    disable=False,
    orientation='horizontal',
    readout=True) 

yr_built = widgets.IntSlider(
    value= 100,
    min = 1900,
    max = 2015,
    step = 1,
    description='Year Built',
    disable=False,
    orientation='horizontal',
    readout=True) 

condition = widgets.Dropdown(
    options=['1', '2', '3', '4', '5'],
    value='2',
    description='Place Condition',
    disabled=False,
)




# Map Att

def update_map( df, limit1, limit2, limit3, limit4, limit5, limit6 ):
    
    houses = df[(df['sqft_living'] <= limit1 ) &
                (df['yr_built'] <= limit5 ) &
                (df['sqft_basement'] <= limit4 ) & 
                (df['price'] <= limit3 ) & 
                (df['bathrooms'] == limit2)][['id', 'lat', 'long', 'price', 'sqft_living', 'bathrooms',
                                              'sqft_basement', 'yr_built', 'condition']]

    fig = px.scatter_mapbox(houses,
                            lat='lat',
                            lon='long',
                            color='price',
                            size='price',
                            color_continuous_scale=px.colors.cyclical.IceFire, 
                            size_max=15,
                            zoom=10 )

    fig.update_layout( mapbox_style='open-street-map' )         
    fig.update_layout( height=600 , margin={'r':0, 't':0, 'l':0, 'b':0} )
    fig.show()

```

```python
widgets.interactive( update_map, df=fixed( df ),
                    limit1=sqft_min, limit2=bath_min, limit3=price_limit,
                    limit4=basement_limit, limit5=yr_built, limit6=condition)
```

```python
# change data format

data['year'] = pd.to_datetime( data['date'] ).dt.strftime( '%Y')
#data['yr_renovated'] = pd.to_datetime( data['yr_renovated'] ).dt.strftime( '%Y')
data['date'] = pd.to_datetime( data['date'] ).dt.strftime( '%Y-%m-%d' )
data['year_week'] = pd.to_datetime( data['date'] ).dt.strftime( '%Y-%U' )


data['is_waterfront'] = data['waterfront'].apply( lambda x: 'yes' if x == 1 else 'no' )



# Widgets to control data

yr_limit = widgets.IntSlider(
    min=1900,
    max=2015,
    value = 50,
    description = 'Disponivel',
    continuous_update=False,
    orientation='horizontal',
    readout=True )

waterfront_bar = widgets.Dropdown(
    options= data['is_waterfront'].unique().tolist(), 
    value='yes',
    description='Water View',
    disable=False )


def update_map( data, limit, waterfront ):
    
    # Filter Data
    
    df = data[(data['yr_renovated'] >= limit) & (data['is_waterfront'] == waterfront) ].copy()
    
    fig = plt.figure( figsize=(21,12))
    specs = gridspec.GridSpec( ncols=2, nrows=2, figure=fig )
    
    ax1 = fig.add_subplot( specs[0, :] ) # First Rows
    ax2 = fig.add_subplot( specs[1, 0] ) # Second Row First Column
    ax3 = fig.add_subplot( specs[1, 1] ) # Second Row Second Column
    
    by_year = df[['id', 'year']].groupby('year').sum().reset_index()
    ax1.bar( by_year['year'], by_year['id'] )
    
    by_day = df[['id', 'date']].groupby('date').mean().reset_index()
    ax2.plot( by_day['date'], by_day['id'] )
    ax2.set_title( 'title: Avg by Day' )
    
    by_week_of_year = df[['id', 'year_week']].groupby('year_week').mean().reset_index()
    ax3.bar( by_week_of_year['year_week'], by_week_of_year['id'] )
    ax3.set_title( 'title: Avg by Week of Year' )
    plt.xticks( rotation=60 )
```

```python
widgets.interactive( update_map, data=fixed( data ), limit=yr_limit, waterfront=waterfront_bar)
```
